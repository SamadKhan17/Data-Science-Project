# K-Means Clustering

```python
df1
```

```python
#Normalization of data

def normalization(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
df1_norm = normalization(df1.iloc[:,:])
df1_norm 
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
```

```python
for i in range(1,11): 
  k = KMeans(n_clusters = i) 
  k.fit(df1_norm.iloc[ : , : ])  
  wcss.append(k.inertia_) 
```

```python
wcss
```

```python
#Elbow Plot

sns.lineplot(x=range(1,11),y = wcss)
plt.title('ELbow/Scree Plot') 
plt.xlabel('K - CLuster Size') 
plt.ylabel('WCSS - Inertia') 

#We can take cluster 4 or 5 by seeing this elbow method as after 4 the WCSS values is not decreasing rapidly
```

```python
kmeans = KMeans(n_clusters = 4) 

predict_clusters = kmeans.fit_predict(df1_norm.iloc[ : , : ])
predict_clusters
```

```python
kmeans.labels_
```

```python
kmeans.inertia_

#WCSS value of Cluser size 4
```

```python
df1['Clusters'] = kmeans.labels_
```

```python
df1
```

```python
#Accuracy of clustering

silhouette_score(df1_norm,df1['Clusters']) 
```

```python
df1[df1['Clusters']==0]
```

# DBSCAN Clustering

```python
df2
```

```python
#Feature Scaling - Standardization

from sklearn.preprocessing import StandardScaler

sc = StandardScaler() 
```

```python
df2_norm = df2.copy()
```

```python
df2_norm.iloc[ : , : ] = sc.fit_transform(df2.iloc[ : , : ])
df2_norm
```

```python
#Estimating minimum points and espilon values by using K neighbors graph

from sklearn.neighbors import NearestNeighbors
 
nn = NearestNeighbors(n_neighbors=20) 

nbrs = nn.fit(df2_norm)
```

```python
dist,indices = nbrs.kneighbors(df2_norm) 
```

```python
#Plotting the graph

dist = np.sort(dist,axis=0)
dist = dist[ : , 1 ]
plt.plot(dist)
plt.show()

#by visualizing this plot we can say Epsilon value has stopped rapidly decreasing from 1, so we can take Epsilon value as 1
```

```python
from sklearn.cluster import DBSCAN 

db = DBSCAN(eps = 1, min_samples = 11)
```

```python
db.fit_predict(df2_norm) 
```

```python
clf = db.fit_predict(df2_norm)
clf
```

```python
db.labels_
```

```python
df2
```

```python
df2['Clusters'] = db.labels_
```

```python
df2
```

```python
df2['Clusters'].unique()
```

```python
#Checking how good the Clustering has been done by using Silhouette Score

silhouette_score(df2_norm,df2['Clusters']) 
```
