# Cutlets.mtv

```python
import pandas as pd
from scipy import stats

import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv('Cutlets.csv')
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
#Boxplots for both Unit A and Unit B

plt.boxplot(df['Unit A'])
plt.title('Unit A')
plt.show
```

```python
plt.boxplot(df['Unit B'])
plt.title('Unit B')
plt.show

#one outlier in Unit B at row 13
```

```python
df['Unit B']<6.2
```

```python
#Plotting distribution plot to understand data distribution

labels=['Unit A','Unit B']
sns.distplot(df['Unit A'],kde=True)
sns.distplot(df['Unit B'],hist=True)
plt.legend(labels)
plt.show()
```

```python
#Hypothesis Testing using t-statistics

stats.ttest_ind(df['Unit A'],df['Unit B'])
```

#Null Hypothesis is Accepted because P-value 0.47 > 0.05(Given Alpha)

#Conclusion: There is no significant difference in the diameter of the cutlet between two units!

# CustomerOrderForm.mtw

```python
df3 = pd.read_csv('Costomer+OrderForm.csv')
```

```python
df3
```

```python
df3.shape
```

```python
df3.isnull().sum()
```

```python
df3.info()
```

```python
from sklearn.preprocessing import LabelEncoder 
le=LabelEncoder()

df3["Phillippines"]=le.fit_transform(df3["Phillippines"])
df3["Indonesia"]=le.fit_transform(df3["Indonesia"])
df3["Malta"]=le.fit_transform(df3["Malta"])
df3["India"]=le.fit_transform(df3["India"])

df3.head()
```

```python
df3.info()
```

```python
#Chi-Sqaure Test

chi2, p, _, _  = stats.chi2_contingency(df3)

print("Chi-Squared Statistic:", chi2)
print("P-value:", p)
```

#Null Hypothesis is Accepted because P-value 1.0 > 0.05(Given Alpha)

#Conclusion: There is no significant differences
