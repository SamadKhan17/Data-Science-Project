```python
mod1 = smf.ols("Profit~Administration + Marketing_Spend",data=df).fit()
mod1.summary()
```

```python
mod2 = smf.ols("Profit~RD_Spend+Administration",data=df).fit()
mod2.summary()
```

```python
mod3 = smf.ols("Profit~RD_Spend+Marketing_Spend",data=df).fit()
mod3.summary()
```

```python
#VIF 

#Calculating colleration between independent variables

#Put the independent variable on Y side(for which you want to calculate VIF) and then put the rest independent variables

m1 = smf.ols('RD_Spend ~ Administration + Marketing_Spend',data=df).fit()
rs_RDS = m1.rsquared
vif_RDS = 1/(1-rs_RDS) 

m2 = smf.ols('Administration ~ RD_Spend + Marketing_Spend',data=df).fit()
rs_ADM = m2.rsquared 
vif_ADM = 1/(1-rs_ADM)

m3 = smf.ols('Marketing_Spend ~ RD_Spend + Administration',data=df).fit()
rs_MAR = m3.rsquared 
vif_MAR = 1/(1-rs_MAR) 
```

```python
print('RD_Spend:', vif_RDS)
print('Administration :',vif_ADM)
print('Marketing_Spend:',vif_MAR)
```

```python
d1 = {"Variables":['RD_Spend', 'Administration', 'Marketing_Spend'], "VIF":[vif_RDS,vif_ADM,vif_MAR]}
d1
```

```python
vif = pd.DataFrame(d1)
vif
```

```python
print('AIC for MOD2 :', mod2.aic)
print('R2 Value of MOD2 :', mod2.rsquared) #RD_Spend + Administration


print('AIC for MOD3 :', mod3.aic) 
print('R2 Value of MOD3 :', mod3.rsquared) #RD_Spend + Marketing_Spend


#We are getting better result using RD_Spend + Marketing_Spend 

#AIC is lower compared to RD_Spend + Administration (as it should be lower for better model)
#R2(square) accuracy is higher compared to RD_Spend + Administration

#In order to remove the multi collinearity we will be building the **mod3** using #RD_Spend + Marketing_Spend independent variables
#and dropping Administration Predictor
```

```python
#test for normality: testing for homoscedasticity

#Quantile-Quantile(QQ) Plot

qqplot = sm.qqplot(model.resid,line="q") 
plt.title('Q-Q Plot of Residuals')
plt.show()

#Most of the residuals/errors(blue dots) should line along with red line, any residuals outside red line is considered outlier
#if maximum amount of residuals/errors(blue dots) lines on the red line that means homoscedasticity is maintained
```

```python
#standardized values for both residuals and fitted values

def standardize1(v):
  eqn = (v - v.mean())/v.std() 
  return eqn
```

```python
sns.scatterplot(x = standardize1(model.fittedvalues),y = standardize1(model.resid)) 
plt.xlabel("Standardized Fitted Values")
plt.ylabel("Standardized Residual Values")
plt.show()
```

```python
#Residuals vs Regressors

#To check how residuals are behaving with RD_Spend regressor(independent variables)

#Considering all independent values of the dataset
#Using it we can identify how the collinearity was present

#Residuals vs Regressor (VOL)
fig = plt.figure(figsize=(14,7))
fig = sm.graphics.plot_regress_exog(model,"RD_Spend",fig=fig) #exog(exogenous) means independent variables(modelname,'predictor',fig for giving size)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Marketing_Spend", fig=fig)
plt.show()
```

```python
fig = plt.figure(figsize=(15,8))
fig = sm.graphics.plot_regress_exog(model, "Administration", fig=fig)
plt.show()
```
