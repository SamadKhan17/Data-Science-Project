```python
df[['job','marital','education','default','housing','loan','contact','month','poutcome','y']] = df[['job','marital','education','default','housing','loan','contact','month','poutcome','y']].apply(lambda x: pd.factorize(x)[0])
```

```python
df
```

```python
df.isna().sum()/len(df)*100
```

```python
#independent variable and dependent variable

x = df.iloc[ : , : 16]
y = df.iloc[ : , -1]
```

```python
x
```

```python
y
```

```python
#build the model and fit the data
from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()
lr.fit(x,y)
```

```python
#prediction

lr.predict(x)
```

```python
ypred = lr.predict(x)
```

```python
#checking accuracy of the model

from sklearn.metrics import confusion_matrix

#building confusion matrix

confusion_matrix(y,ypred)
```

```python
from sklearn.metrics import classification_report #use classification report to check for accuracy

print(classification_report(y,ypred))

#89% accuracy of the model
```

```python
from sklearn.metrics import accuracy_score #can calculate accuracy simply by using accracy_score

accuracy_score(y,ypred)

#88.7% accuracy almost similar to classification report accuracy
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
##AUC score

auc = roc_auc_score(y,lr.predict(x))
auc                                   
```

```python
fpr,tpr,thresholds = roc_curve(y,lr.predict_proba(x)[:,1]) 
```

```python
#ROC curve visualization

plt.plot(fpr,tpr,color = 'red') 
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.show()
```

```python
plt.plot(fpr,tpr,color = 'red')
plt.plot([0,1],[0,1]) 
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('Receiver Operating Characteristic')
plt.show()
```
